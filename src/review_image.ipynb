```python
import os
import random
import shutil

img_dir = "../data/images01"
save_dir = "../data/images010"
for folder in os.listdir(img_dir):
    folder_path = os.path.join(img_dir, folder)
    img_list = []
    for img in os.listdir(folder_path):
        img_path = os.path.join(folder_path, img)
        img_list.append(img_path)
    random.shuffle(img_list)
    save_folder = os.path.join(save_dir, folder)
    os.makedirs(save_folder, exist_ok=True)
    for i, img_path in enumerate(img_list):
        # Tạo tên mới cho tệp
        new_name = "{}{}".format(i + 1, os.path.splitext(img_path)[-1])
        # Tạo đường dẫn mới cho tệp
        new_path = os.path.join(save_folder, new_name)
        # Copy qua thư mục mới
        shutil.copy(img_path, new_path)
print("Xong.")
```

```python
img_dir = "../data/images01"
```

```python
import os

categories = os.listdir(img_dir)
print(categories)
```

```python
import matplotlib.pyplot as plt
for category in categories:

    fig, _ = plt.subplots(3,4)
    fig.suptitle(category)

    for k, v in enumerate(os.listdir(os.path.join(img_dir, category))[:12]):
        img = plt.imread(os.path.join(img_dir, category, v))
        plt.subplot(3, 4, k+1)
        plt.axis('off')
        plt.imshow(img)
    plt.show()
```

```python
for category in categories:

    shape0 = []
    shape1 = []

    for files in os.listdir(os.path.join(img_dir, category)):
        files_path = os.path.join(img_dir, category, files)

        shape0.append(plt.imread(files_path).shape[0])
        shape1.append(plt.imread(files_path).shape[1])

    print('{:5s} => height min: {:5d},    width min: {:5d}'.format(category, min(shape0), min(shape1)))
    print('{:5s} => height max: {:5d},    width max: {:5d}'.format(category, max(shape0), max(shape1)))
    print()
```
